# **1. Line Plot Assessment (Matplotlib)**

```python
# Task:
# Create a line graph showing monthly sales:
# Required:
# Plot line graph
# Label axes
# Add title
# Add marker for each point
# Change line color

import matplotlib.pyplot as plt
import seaborn as sns

months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [120, 150, 170, 160, 180, 200]
plt.plot(months,sales,marker='o',color="Pink")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.show()
```

# **2. Bar Chart Assessment (Matplotlib)**

```python
# Task:
# Plot a bar chart for student marks:
# Required:
# Horizontal bar chart
# Add different colors
# Display values above bars

students = ["A","B","C","D","E"]
marks = [78, 92, 67, 85, 90]
plt.bar(students,marks,color=["Maroon","Brown","Red","Orange","Yellow"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students Marks")
plt.show()
```

# **3. Scatter Plot Assessment (Matplotlib)**

```python
# Task:
# Visualize relationship between hours studied vs marks:
# Required:
# Scatter plot
# Add trendline
# Use different marker style

hours = [1,2,3,4,5,6,7]
marks = [55,60,62,68,72,78,85]
plt.scatter(hours,marks,marker="*",color="Maroon")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Hours studied vs Marks")
plt.show()
```

# **4. Histogram Assessment (Seaborn)**

```python
# Task:
# Given a list of ages:
# Required:
# Plot a histogram with seaborn
# Add KDE curve (kde=True)
# Use 10 bins

import seaborn as sns

ages = [23,25,21,30,22,28,26,32,35,29,27,24]
sns.histplot(ages,bins=10,kde=True,color="green")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.title("Frequency of Ages using Seaborn")
plt.show()
```

# **5. Boxplot Assessment (Seaborn)**

```python
# Task:
# Plot salary distribution:
# Required:
# Create a boxplot
# Identify outliers (if any)
# Explain what median and whiskers mean

salary = [30000,35000,38000,45000,50000,52000,60000,62000,70000,72000]
sns.boxplot(data=salary,color="Black")
plt.ylabel("Salary")
plt.show()
```

# **6. Countplot Assessment (Seaborn)**

```python
# Task:
# Visualize employee departments:
# Required:
# Create countplot
# Change color palette
# Add title
departments = ["HR","IT","IT","Finance","HR","Sales","IT","Sales","Finance","HR"]
sns.countplot(x=departments,palette="viridis")
plt.title("Employee Count by Department")
plt.show()
```

# **7. Heatmap Assessment (Seaborn)**

```python
# Task:
# Plot heatmap of correlation matrix:
import seaborn as sns
import pandas as pd
df = sns.load_dataset("tips")   # built-in dataset
# Required:
# Compute df.corr()
# Plot heatmap
# Annotate the values (annot=True)

corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
sns.heatmap(corr_matrix, annot=True, cmap="rocket", linewidths=0.5)
plt.title("Correlation Heatmap - Tips Dataset")
plt.show()
```

# **8. Pie Chart Assessment (Matplotlib)**

```python
# Task:
# Visualize market share:
# Required:
# Create pie chart
# Show percentages
# Add explode effect to 1st slice

brands = ["Apple","Samsung","Xiaomi","OPPO"]
share = [40,30,20,10]

explode = [0.1, 0, 0, 0]
plt.pie(share, labels=brands, autopct="%1.1f%%", explode=explode, shadow=True)
plt.title("Market Share Analysis")
plt.show()
```

# **9. Pairplot Assessment (Seaborn)**

```python
# Task:
# Use the iris dataset:
df = sns.load_dataset("iris")
# Required:
# Create pairplot
# Color by species
# Explain one insight from the graph

sns.pairplot(df, hue="species")
plt.show()
```

# **10. Subplot Assessment (Matplotlib)**

```python
# Task:
# Using:
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
# Required:
# Create a figure with 2 subplots:
# 1st subplot  Line chart
# 2nd subplot  Scatter plot
# Add titles to each
# Add overall figure title

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(x, y1)
ax1.set_title("Line Chart")
ax2.scatter(x, y2)
ax2.set_title("Scatter Plot")
fig.suptitle("Comparison of Line and Scatter Plots", fontsize=14)
plt.show()
```
