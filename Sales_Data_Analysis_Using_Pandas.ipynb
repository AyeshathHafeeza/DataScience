# **Mini Project: "Sales Data Analysis using Pandas"**

```python
import pandas as pd

# Load and inspect the data
df=pd.read_csv("/content/sales - pandas 1.csv")
print(df.head())
```

```python
# identifying if there is missing values
df.isnull().sum()
```

```python
#remove missing values
df = df.dropna()
df.isnull().sum() # as there is no missing values
```

```python
# Displays first 5 rows.
df.head()
```

```python
#Count duplicates
df.duplicated().sum()
```

```python
df.head()
```

```python
#Remove Duplicates
df = df.drop_duplicates()
df.head()
```

```python
#calculate Revenue
df['Revenue']=df['Price']*df['Quantity']
print(df['Revenue'])

#Total Price
total_price=df['Revenue'].sum()
print("Total Price:",total_price)
```

```python
#Top customers
df.groupby('Customer Name')['Revenue'].sum().sort_values(ascending=False).head(10)

```

```python
#Top products
df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
#Total revenue by customers
df.groupby("Customer Name")["Revenue"].sum()
```

```python
#Total revenue by Product
df.groupby("Product")["Revenue"].sum()
```

```python
#Total revenue by Location
df.groupby("Location")["Revenue"].sum()
```

```python
#Total Quantity Sold per Product
df.groupby("Product")["Quantity"].sum()
```

```python
#Multi-level Grouping (Customer + Product)
df.groupby(["Customer Name","Product"])["Revenue"].sum()
```

```python
#Number of orders per customer
df.groupby("Customer Name").size()
```

```python
#Average order value per customer
df.groupby("Customer Name")["Revenue"].mean()
```

```python
#Top N products
df.groupby("Product")["Revenue"].sum().nlargest(5)
```

```python
#sort for highest sales
df.groupby("Product")["Price"].sum()
```

```python
#Filter with query()

#Filter rows by Customer
df.query("`Customer Name` == 'John'")
```

```python
#Filter rows by Product
df.query("Product == 'Laptop'")
```

```python
#Filter rows with multiple conditions
df.query("Product == 'Laptop' and Location == 'Mumbai'")
```

```python
#Filter by Quantity greater than a value
df.query("Quantity > 1")
```

```python
#Filter by Price less than a value
df.query("Price < 20000")
```
