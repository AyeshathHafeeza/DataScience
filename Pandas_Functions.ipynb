# **Pandas Functions**

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
# Displays the first 5 rows of the DataFrame.
df.head()
```

```python
# Summarizes your DataFrameâ€”columns, data types, and missing values.
df.info()
```

```python
# Provides summary statistics for all numeric columns.
df.describe()
```

```python
# Counts missing (null) values in each column.
df.isnull().sum()
```

```python
# Replaces missing values with a specific value.
df.fillna(2)
```

```python
# Groups the data by a column to perform aggregation (sum, mean, count).
df.groupby('longitude')['latitude'].sum()
```

```python
# Counts how often each value appears in a column.
df['longitude'].value_counts()
```

```python
# Sorts the rows based on values in a specific column.
df.sort_values(by='housing_median_age')
```

```python
# Access specific rows by label (loc) or index (iloc).
df.loc[3]
```

```python
# Applies a function to each row or column.
df['Discounted Price'] = df['median_income'].apply(lambda x: x * 0.9)
print(df['Discounted Price'])
```

```python
# Removes rows with missing values.
df.dropna()
```

```python
# Identifies repeated rows.
df.duplicated()
```

```python
# Removes repeated rows
df.drop_duplicates()
```

```python
# Shows column names.
df.columns
```

```python
# Gives dimensions (rows, columns).
df.shape
```

```python
# Returns all unique values.
df['latitude'].unique()
```

```python
# Counts how many unique values.
df['latitude'].nunique()
```

```python
# Filters rows with a SQL-like expression (more readable than complex [] filtering).
df.query("total_rooms>2000")
```
