# **NumPy Basic Functions**

---



```python
import numpy as np
```

```python
# Creates a NumPy array from a list or tuple.
arr = np.array([1, 2, 3])
print(arr)
```

```python
# Creates a range of values with a step.
np.arange(1, 10, 2)
```

```python
# Generates evenly spaced values between a start and end.
np.linspace(0, 1, 5)
```

```python
# Creates an array of zeros.
np.zeros((3,2))
```

```python
# Creates an array of ones.
np.ones(5)
```

```python
# Generates an identity matrix
np.eye(4)
```

```python
#Changes array shape without changing the data.
arr=np.array([4,5,3,2,6,7])
arr.reshape(2, 3)
```

```python
# Adds all elements.
np.sum(arr)
```

```python
# Calculates average value.
np.mean(arr)
```

```python
# Computes standard deviation.
np.std(arr)
```

```python
# Finds smallest
np.min(arr)
```

```python
# Finds largest
np.max(arr)
```

```python
# Generates random numbers between 0 and 1.
np.random.rand(3)
```

```python
# Random integers in a specified range.
np.random.randint(1, 100)
```

```python
# Combines arrays.
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
np.concatenate([a, b])
```

```python
# Returns unique elements.
np.unique(arr)
```

```python
# Sorts an array.
np.sort(arr)
```

```python
# Access specific rows, columns and slices.

arr[:]
```

```python
# Returns positions where condition is true.
np.where(arr > 6)
```

```python
# Matrix multiplication / dot product.
np.dot(a, b)
```

```python
# Swaps rows and columns
arr.T
```
